```python
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
import lyricsgenius
import pandas as pd
import requests

# Spotify API setup
client_credentials_manager = SpotifyClientCredentials(client_id='96139f35ac8f4d789adf48fd8f218f81', client_secret='4ce0a0ac43144e1c9607ad24942465cd')
sp = spotipy.Spotify(client_credentials_manager=client_credentials_manager)

# Genius API setup
genius = lyricsgenius.Genius('n5GrVP6_8wLjSmD0IV6GIfe3EU1mnZ1wu821E26iN4C-Eal-XgBeL7fL5bpNWVpA')

# Lists to store data
artist_name, track_name, track_popularity, artist_id, track_id, lyrics = [], [], [], [], [], []

# Function to retrieve lyrics from Genius
def get_lyrics(track_name, artist_name, lang='ko'):
    lyrics_url = f'https://api.lyrics.ovh/v1/{artist_name}/{track_name}?lang={lang}'
    lyrics_response = requests.get(lyrics_url)
    lyrics_data = lyrics_response.json()
    return lyrics_data.get('lyrics', 'Lyrics not available.')

# Get the playlist ID for the "대한민국 차트 50" playlist
playlist_id = "37i9dQZEVXbNxXF4SkHj9F"

# Retrieve playlist tracks
charts = sp.playlist(playlist_id)

# Loop through playlist tracks and get lyrics
for track in charts['tracks']['items']:
    artist_name.append(track['track']['artists'][0]['name'])
    artist_id.append(track['track']['artists'][0]['id'])
    track_name.append(track['track']['name'])
    track_id.append(track['track']['id'])
    track_popularity.append(track['track']['popularity'])

    # Retrieve lyrics from Genius
    try:
        song = genius.search_song(track['track']['name'], artist_name[-1])
        lyrics.append(song.lyrics)
    except AttributeError:
        lyrics.append(get_lyrics(track['track']['name'], artist_name[-1]))

# Create DataFrame
df = pd.DataFrame({
    'Artist': artist_name,
    'Artist_ID': artist_id,
    'Track': track_name,
    'Track_ID': track_id,
    'Popularity': track_popularity,
    'Lyrics': lyrics
})

# Save DataFrame to a CSV file
df.to_csv('combined_data_with_lyrics.csv', index=False)

```
