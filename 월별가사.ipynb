```python
raw_df['Month'] = raw_df['Month'].astype(str)
raw_df['year'] = raw_df['Month'].str[:4].astype(int)
raw_df['month'] = raw_df['Month'].str[4:].astype(int)

raw_df.head()
```

```python
import pandas as pd
import re
from konlpy.tag import Okt
from collections import Counter
import matplotlib.pyplot as plt
import matplotlib
from matplotlib import font_manager, rc
from wordcloud import WordCloud
import matplotlib.font_manager as fm

# Assuming df is your list of DataFrames
for i in range(len(df)):
    # 중복 제거
    df[i] = df[i].drop_duplicates(subset=['가사'])

    # 불용어 파일 읽기
    with open('stopwords.txt', 'r', encoding='utf-8') as file:
        stop_words = file.readlines()
        stop_words = [word.strip() for word in stop_words]

    okt = Okt()

    # 불용어 제거
    def extract_nouns(text):
        nouns = okt.nouns(text)
        return [word for word in nouns if word not in stop_words and len(word) > 1]

    df[i]['가사_명사'] = df[i]['가사'].astype(str).apply(extract_nouns)

    all_nouns = df[i]['가사_명사'].sum()

    # Counter를 사용하여 명사 빈도 계산
    count_nouns = Counter(all_nouns)
```

```python
font_path= "/Users/yoosukyeong/Library/Fonts/LeferiPointBlack.ttf"
font_name = fm.FontProperties(fname=font_path).get_name()
#상위 50 출력
fig, axes = plt.subplots(3, 4, figsize=(15, 10))

for i in range(len(df)):
    # Your existing code for processing each DataFrame
    top_nouns = dict(count_nouns.most_common(50))
    
    wc = WordCloud(font_path, background_color='white', \
                   width=300, height=300, max_words=100, max_font_size=100)
    wc.generate_from_frequencies(top_nouns)

    # Display the word cloud in the i-th subplot
    row = i // 4
    col = i % 4
    axes[row, col].imshow(wc, interpolation='bilinear')
    axes[row, col].axis('off')
    axes[row, col].set_title(f'Month {i + 1}')

# Adjust layout
plt.tight_layout()
plt.show()
```
