```python
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
import lyricsgenius
import pandas as pd

# Spotify API setup
client_credentials_manager = SpotifyClientCredentials(client_id='96139f35ac8f4d789adf48fd8f218f81', client_secret='4ce0a0ac43144e1c9607ad24942465cd')
sp = spotipy.Spotify(client_credentials_manager=client_credentials_manager)

# Genius API setup
genius = lyricsgenius.Genius('n5GrVP6_8wLjSmD0IV6GIfe3EU1mnZ1wu821E26iN4C-Eal-XgBeL7fL5bpNWVpA')

# Lists to store data
artist_name = []
track_name = []
track_popularity = []
artist_id = []
track_id = []
lyrics = []

# Loop through Spotify tracks
for i in range(0, 100, 50):
    track_results = sp.search(q='year:2022', type='track', limit=50, offset=i, market='KR')
    
    for i, t in enumerate(track_results['tracks']['items']):
        artist_name.append(t['artists'][0]['name'])
        artist_id.append(t['artists'][0]['id'])
        track_name.append(t['name'])
        track_id.append(t['id'])
        track_popularity.append(t['popularity'])
        
        # Retrieve lyrics from Genius
        try:
            song = genius.search_song(t['name'], artist_name[-1])
            lyrics.append(song.lyrics)
        except AttributeError:
            lyrics.append(None)  # If lyrics are not found, store None

# Create DataFrame
df = pd.DataFrame({
    'Artist': artist_name,
    'Artist_ID': artist_id,
    'Track': track_name,
    'Track_ID': track_id,
    'Popularity': track_popularity,
    'Lyrics': lyrics
})

# Save DataFrame to a CSV file
df.to_csv('spotify_data_with_lyrics.csv', index=False)
```
