```python
import pandas as pd
import re
from konlpy.tag import Okt
from collections import Counter
import matplotlib.pyplot as plt
import matplotlib
from matplotlib import font_manager, rc
from wordcloud import WordCloud
import matplotlib.font_manager as fm
import seaborn as sns

raw_df = pd.read_csv(r'music_최종.csv', parse_dates=['발매일'], encoding='cp949')
raw_df.head()
```

```python
# '제목' 컬럼에서 괄호와 그 안의 내용 제거
raw_df['제목_가공'] = raw_df['제목'].replace(to_replace='\([^)]*\)', value='', regex=True)
raw_df.head()
```

```python
df = raw_df[raw_df['순위'] == 1]
df.head()
```

```python
# 중복 제거
df = df.drop_duplicates(subset=['제목_가공'])

# 불용어 파일 읽기
with open('stopwords.txt', 'r', encoding='utf-8') as file:
    stop_words = file.readlines()
    stop_words = [word.strip() for word in stop_words]

okt = Okt()

#불용어 제거
def extract_nouns(text):
    nouns = okt.nouns(text)
    return [word for word in nouns if word not in stop_words and len(word) > 1]

df['제목_명사'] = df['제목_가공'].astype(str).apply(extract_nouns)

all_nouns = df['제목_명사'].sum()

# Counter를 사용하여 명사 빈도 계산
count_nouns = Counter(all_nouns)

font_path= "/Users/yoosukyeong/Library/Fonts/LeferiPointBlack.ttf"
font_name = fm.FontProperties(fname=font_path).get_name()

top_nouns = dict(count_nouns.most_common(10))
top_nouns_df = pd.DataFrame(top_nouns.items(), columns=['명사', '빈도'])

plt.rc('font', family=font_name)
plt.figure(figsize=(12, 8))
sns.barplot(x='명사', y='빈도', data=top_nouns_df, palette='viridis')
plt.title('1위 곡 제목 명사 빈도')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 10))
plt.pie(top_nouns_df['빈도'], labels=top_nouns_df['명사'], autopct='%1.1f%%', startangle=140, colors=sns.color_palette('viridis'))
plt.title('1위 곡 제목 명사 빈도')
plt.show()
```
