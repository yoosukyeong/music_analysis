```python
import pandas as pd

df = pd.read_csv('combined_data_with_lyrics.csv')
df.head()
```

```python
df['Lyrics'] = df['Lyrics'].apply(lambda x: '\n'.join(x.split('\n')[1:]))
df['Lyrics'] = df['Lyrics'].str.replace('\n', '')
df.head()
```

```python
import re

df['Lyrics'] = df['Lyrics'].apply(lambda x: re.sub(r'\[.*?\]', '', x))
df.head()
```
