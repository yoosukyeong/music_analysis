```python
import pandas as pd
import numpy as np

raw_df = pd.read_csv(r'music_최종.csv', parse_dates=['발매일'], encoding='cp949')
raw_df.head()
```

```python
# '제목' 컬럼에서 괄호와 그 안의 내용 제거
raw_df['제목_가공'] = raw_df['제목'].replace(to_replace='\([^)]*\)', value='', regex=True)
raw_df.head()
```

```python
raw_df['Month'] = raw_df['Month'].astype(str)
raw_df['year'] = raw_df['Month'].str[:4].astype(int)
raw_df['month'] = raw_df['Month'].str[4:].astype(int)

raw_df.head()
```

```python
# Month열 삭제
raw_df = raw_df.drop('Month', axis=1)

raw_df.info()
```

```python
raw_df['season'] = np.where(raw_df['month'].isin([3, 4, 5]), '봄',
                   np.where(raw_df['month'].isin([6, 7, 8]), '여름',
                   np.where(raw_df['month'].isin([9, 10, 11]), '가을', '겨울')))
raw_df.head()
```

```python
spring = raw_df[raw_df['season'] == '봄']
summer = raw_df[raw_df['season'] == '여름']
fall = raw_df[raw_df['season'] == '가을']
winter = raw_df[raw_df['season'] == '겨울']
```

```python
spring.head()
```

```python
# 봄

import pandas as pd
import re
from konlpy.tag import Okt
from collections import Counter
import matplotlib.pyplot as plt
import matplotlib
from matplotlib import font_manager, rc
from wordcloud import WordCloud
import matplotlib.font_manager as fm
import seaborn as sns

# 중복 제거
spring = spring.drop_duplicates(subset=['제목_가공'])

# 불용어 파일 읽기
with open('stopwords.txt', 'r', encoding='utf-8') as file:
    stop_words = file.readlines()
    stop_words = [word.strip() for word in stop_words]

okt = Okt()

#불용어 제거
def extract_nouns(text):
    nouns = okt.nouns(text)
    return [word for word in nouns if word not in stop_words and len(word) > 1]

spring['제목_명사'] = spring['제목_가공'].astype(str).apply(extract_nouns)

all_nouns_spring = spring['제목_명사'].sum()

# Counter를 사용하여 명사 빈도 계산
count_nouns_spring = Counter(all_nouns_spring)

font_path= "/Users/yoosukyeong/Library/Fonts/LeferiPointBlack.ttf"
font_name = fm.FontProperties(fname=font_path).get_name()

# 상위 10위 제목
top_nouns = dict(count_nouns_spring.most_common(10))
top_nouns_df = pd.DataFrame(top_nouns.items(), columns=['명사', '빈도'])

plt.rc('font', family=font_name)
plt.figure(figsize=(12, 8))
sns.barplot(x='명사', y='빈도', data=top_nouns_df, palette='viridis')
plt.title('봄 상위 10개 제목 명사 빈도')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```
